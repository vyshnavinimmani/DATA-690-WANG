```
# using integers_list to store values

integers_list = []
## using 'for' loop for entering 10 input values
for i in range(1,11):
     # asking user to enter integers
     num = int(input('Enter an Integer:'))
     # Printing the user provided input values
     print(f"Integer #{i}:",num)
      # appending the provided input values in the list1
     integers_list.append(num)


#Printing the list of values
print("The entered values in integers_list are:",integers_list)







    
```

```
#Printing the minimum and maximum of integers_list using min and max 
print("Minimum value of the list is:",min(integers_list))
print("Maximum value of the list is:",max(integers_list))
#storing range1 as range of the list of provided values
range1 = (min(integers_list),max(integers_list))
print("Range of the integers_list is:",range1)
```

```
#Printing mean as sum of values / length of list
print("Mean of the integer_list is:",sum(integers_list)/len(integers_list))
```

```
#calculating variance with 2 different formulas 
##sample variance is divided by (len(integers_list)-1) and population variance by (len(integers_list))

sam_var = sum([(xi - (sum(integers_list)/len(integers_list)))**2 for xi in integers_list]) / (len(integers_list) -1)

popu_var = sum([(xi - (sum(integers_list)/len(integers_list)))**2 for xi in integers_list]) / (len(integers_list))

print("Sample Varience of the list is:",sam_var)
print("Population Varience of the list is:",popu_var)
```

```
#Standard deviation is square root of variance

print("Standard deviation of the list is:",(sam_var)**(1/2))
```
