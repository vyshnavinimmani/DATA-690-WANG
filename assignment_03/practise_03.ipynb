```
import random
## to print matrix of order 10 by 10 and place * if the number is odd

for i in range(0,10):
    
    for y in range(0,10):
      x = random.randint(0,9)
      if ((x % 2) == 0):
        print(x,end = " ")
      else:
        print("*",end = ' ')
    print("\n")
```

```
import random
## calculate and print total of each row in a 10 by 10 matrix
for i in range(0,10):
    sum=0
    for y in range(0,10):
        x=random.randint(0,9)
        print(x,end = " ")
        sum=sum+x
        if y == 9:
            print('* {total}'.format(total=sum))
    print("\n")
    
```
