```python
weather = pandas.read_csv('weather.csv')
weather.head()
```

```python
weather_melt = weather.melt(
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)
weather_tidy.head()
```

```python
weather_tidy.reset_index()
```

```python
billboard_songs = billboard_melt[['year', 'artist', 'track', 'time']]
billboard_songs.head()
```

```python
billboard_ratings = billboard_melt.merge(
    billboard_songs, on=['year', 'artist', 'track', 'time']
)
billboard_ratings.head()
```

```python
billboard_ratings=billboard_ratings[['id', 'date.entered', 'week', 'rating']]
billboard_ratings.head()
```

Learned from the video are:
1. With df.values we can get numpy array version of our dataset allowing you to load data into another program.
2. loc[2] and iloc[2] are different.iloc[2] gives index value with 2 and loc[2] gives the second label.
3. In Python 1_000 same as 1000.
4. pd.melt() changes wide data to long data . converts multiple columns to single column.
5. 'expand' function of split().
6. 'reset_index()' sets the index.
7. 'drop_duplicates()' drops all duplicates.
8. Pivot can be used as opposite of melt.
